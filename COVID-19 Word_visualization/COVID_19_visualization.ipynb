**欄位名稱**

 ObservationDate :日期<br>
 Province/State :區域<br>
 Country/Region	:國家<br>
 Last Update	:最後更改時間<br>
 Confirmed    :確診人數<br>
 Deaths:死亡人數<br>
 Recovered:治癒人數<br>
 Active_case:目前感染人數<br>

 Active_case = Confirmed - Deaths - Recovered


```
# 利用 PANDAS 讀入DATA

data = pd.read_csv("covid_19_data.csv")
data.head()
```

```
# 檢視資料集
data.info()
```

```
#缺失值在資料中的百分比
NAN = [(c, data[c].isna().mean()*100) for c in data]
NAN = pd.DataFrame(NAN, columns=["column_name", "percentage"])
NAN
```

```
#缺失的區域值替代為"Unknown"
data["Province/State"]= data["Province/State"].fillna("Unknown")
#
data["Country/Region"] = data["Country/Region"].replace("Mainland China", "China")
data.head(10)
```

```
#確認無缺失值
data.info()
```

```
# 原始資料集並無感染的欄位, 利用資料處理新增一項目 "Active_case:感染人數"
# Active_case = Confirmed - Deaths - Recovered

data['Active_case'] = data['Confirmed'] - data['Deaths'] - data['Recovered']
data.head()
```

#2020年1/22~ 11/15總人數統計

```
max(data['ObservationDate'])
```

```
Data = data[data['ObservationDate'] == max(data['ObservationDate'])].reset_index()
Data.head()
```

```
#最終統計人數
Data_world = Data.groupby(["ObservationDate"])[["Confirmed","Active_case","Recovered","Deaths"]].sum().reset_index()
Data_world
```

## 繪製圓餅圖統計死亡、感染、治癒百分比

```
labels = ["感染","治癒","死亡"]
#計算數字
values = Data_world.loc[0, ["Active_case","Recovered","Deaths"]]
print(values)

fig = px.pie(Data_world, values=values, names=labels,color_discrete_sequence=['rgb(77,146,33)','rgb(69,144,185)','rgb(77,77,77)'],hole=0.7)
fig.update_layout(title='總共確診人數 :{}人 '.format(str(int(Data_world["Confirmed"][0]))))
fig.show()
```

##建立時間線

```
data_over_time= data.groupby(["ObservationDate"])[["Confirmed","Active_case","Recovered","Deaths"]].sum().reset_index()
data_over_time.head()
#print(data_over_time)
```

```
fig = go.Figure()
fig.add_trace(go.Scatter(x=data_over_time.index, y=data_over_time['Confirmed'],
                    mode='lines',
                    name='確診人數'))

fig.add_trace(go.Scatter(x=data_over_time.index, y=data_over_time['Recovered'],
                    mode='lines',
                    name='治癒人數'))
fig.add_trace(go.Scatter(x=data_over_time.index, y=data_over_time['Deaths'],
                    mode='lines',
                    name='死亡人數'))

#fig.add_trace(go.Scatter(x=data_over_time.index, y=data_over_time['Active_case'],
#                    mode='lines',
#                    name='感染人數'))
fig.update_layout(
    title='世界疫情統計',
        template='plotly_dark',
      yaxis_title="Number of people",
    xaxis_title="Days",

)

fig.show()
```

#2020年1/22~ 11/15各國人數統計

##統計各國人數

```
Data_per_country = Data.groupby(["Country/Region"])["Confirmed","Active_case","Recovered","Deaths"].sum().reset_index().sort_values("Confirmed",ascending=False).reset_index(drop=True)
#print("各國:",Data_per_country["Country/Region"].unique())
print("總共有:{}個國家".format(len(Data_per_country)))
Data_per_country.head()
```

```
headerColor = 'grey'
rowEvenColor = 'lightgrey'
rowOddColor = 'white'

fig = go.Figure(data=[go.Table(
  header=dict(
    values=['<b>國家</b>','<b>確診人數</b>','<b>目前感染人數</b>','<b>治癒人數</b>','<b>死亡人數</b>'],
    line_color='darkslategray',
    fill_color=headerColor,
    align=['left','center'],
      
    font=dict(color='white', size=12)#字體設定
  ),
  cells=dict(
    values=[
      Data_per_country['Country/Region'],
      Data_per_country['Confirmed'],
      Data_per_country['Active_case'],
      Data_per_country['Recovered'],
      Data_per_country['Deaths'],

      ],
    line_color='darkslategray',
    # 2-D list of colors for alternating rows
    fill_color = [[rowOddColor,rowEvenColor,rowOddColor, rowEvenColor,rowOddColor]*len(Data_per_country)],
    align = ['left', 'center'],#值的位置
    font = dict(color = 'darkslategray', size = 11)
    ))
])
fig.update_layout(
    title='各國疫情統計',
)
fig.show()
```

##各國確診人數

##世界地圖上確診分佈

```
fig = px.choropleth(Data_per_country, locations=Data_per_country['Country/Region'],
                    color=Data_per_country['Confirmed'],locationmode='country names', 
                    hover_name=Data_per_country['Country/Region'],  #每個國家的標籤
                    color_continuous_scale=px.colors.sequential.Tealgrn,#內建連續顏色https://plotly.com/python/builtin-colorscales/
                    template='plotly_dark', )#三个内置的 Plotly 主题可以使用 plotly、plotly_white、plotly_dark 
fig.update_layout(
    title='確診國家',
)
fig.show()
```
